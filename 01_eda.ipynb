```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
import sys

from data.load_data import load_interactions

project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(project_root / "src"))
data_dir = project_root / "data" / "raw"
interactions = load_interactions(str(data_dir))

interactions.head()
```

```python
#Basic Stats
n_users = interactions["user_id"].nunique()
n_items = interactions["item_id"].nunique()
n_interactions = len(interactions)
print (f"Users: {n_users}")
print (f"Items: {n_items}")
print (f"Interactions: {n_interactions}")
print (f"Average interactions per user: {n_interactions / n_users:.2f}")
print (f"Average interactions per item: {n_interactions / n_items:.2f}")
```

```python
#User Activity Distribution
user_counts = interactions["user_id"].value_counts()
user_counts.describe()
user_counts.hist(bins = 50)
plt.xlabel("Interactions per user")
plt.ylabel("Count of users")
plt.title("Distribution of interactions vs count of users")
plt.show()
```

```python
#Item popularity distribution
item_counts = interactions["item_id"].value_counts()
item_counts.describe()
item_counts.hist(bins = 50)
plt.xlabel("Interactions per item")
plt.ylabel("Count of items")
plt.title("Distribution of interactions vs count of items")
plt.show()
```
